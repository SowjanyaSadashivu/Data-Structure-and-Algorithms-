```python
class Node:
    def __init__(self, data):
        self.data  = data
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def insert_beggining(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node

    def insert_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def insert_list(self, data_list):
        #self.head = None
        for data in data_list:
            self.insert_end(data)
    
    def insert_at(self, index, data):
        if index < 0 or index >= int(self.get_length()):
            raise Exception('Invalid index')
        
        if index == 0:
            self.insert_node_at_beggining(data)
            return

        count = 0
        current = self.head
        while current:
            if count == index - 1:
                new_node = Node(data)
                new_node.next = current.next
                current.next = new_node
                break
            current = current.next
            count += 1

    def insert_after(self, after, data): 
        if self.head is None:
            return

        if self.head.data == after:
            new_node = Node(data)
            new_node.next = self.head.next
            self.head.next =  new_node
            return
             
        current = self.head
        while current:
            if current.data == after:
                new_node = Node(data)
                new_node.next = current.next
                current.next = new_node
                break
            current = current.next 

    def get_length(self):
        count = 0
        if self.head is None:
            return
        
        current = self.head
        while current:
            count+=1
            current = current.next
        #print(f'length of linked list is : {count}')
        return count

    def remove_at(self, index):
        if index < 0 or index >= int(self.get_length()):
            raise Exception('Invalid index')

        if index == 0:
            self.head = self.head.next
            return
        
        count = 0
        current = self.head
        while current:
            if count == index - 1:
                current.next = current.next.next
                break
            current = current.next
            count += 1

    def delete_node(self, data):
        if self.head is None:
            print('No values in linked list')
            return
        
        if self.head == data:
            self.head = self.head.next
            return

        current = self.head

        while current.next:
            if current.next.data == data:
                current.next = current.next.next
                return

            current = current.next
        
        print(f'{data} not found in linked list')

    def display(self):
        if self.head is None:
            print('No nodes in the linked list')
        i = ''
        current = self.head
        while current:
            i += str(current.data) + ' --> '
            current = current.next

        print(i)    
```

### Double Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
```

```python
class DoubleList:
    def __init__(self):
        self.head = None

    def insert_data(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node
        new_node.prev = current
    
    def inset_list(self, data):
        for i in data:
            self.insert_data(i)

    def print_forward(self):
        if self.head is None:
            print('There are no data in the list')
            return
        
        current = self.head
        i = ''
        values = []
        while current:
            i += str(current.data) + '-->'
            values.append(current.data)
            current = current.next
        print("forward: ", i)
        return values


    def print_backwards(self, data):
        arr = ''
        for i in data[::-1]:
            arr += str(i) + '-->'

        print('Backward: ', arr)

    def get_last(self):
        current = self.head
        while current.next:
            current = current.next
        return current


    def print_backs(self):
        if self.head is None:
            print('No data in list')
            return
        
        current = self.get_last()
        back = ''
        while current:
            back += str(current.data) + '-->'
            current = current.prev
        print('List backwards in other method, without [::-1] is: ', back)

```

```python
#creating doublelist obj
dl = DoubleList()
#first method : insert values
dl.insert_data(10)
dl.inset_list([12, 32, 34, 24])
data = dl.print_forward()
dl.print_backwards(data)
dl.print_backs()
```
